```python
%matplotlib notebook
from spatialmath import *
from spatialmath.base import *
import roboticstoolbox as rtb
from math import pi
import numpy as np
```

```python
l1 = 1
l2 = 1
q1 = pi/6
q2 = -pi/6
```

```python
S0 = SE3()
S0
```

```python
A01 = SE3.Rz(q1)*SE3.Tx(l1)
A01
```

```python
S1 = S0 @ A01
S1
```

```python
A12 = SE3.Rz(q2)*SE3.Tx(l2)
A12
```

```python
S2 = S1 @ A12
S2
```

```python
trplot(S0.A, dims=[-1,5,-1,5,-1,5])
trplot(S1.A)
trplot(S2.A)
```

```python
tranimate(S2.A,start=S1.A, frame='', dims=[0, 5], nframes=10, interval = 200)
```

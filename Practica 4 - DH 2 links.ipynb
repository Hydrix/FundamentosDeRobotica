```python
%matplotlib notebook 
# Para tener graficos interactivos
from spatialmath import * 
# Para usar SE3 de manera directa
from spatialmath.base import * 
# Para usar trplot de manera directa
from math import pi 
# Para definir los angulos
```

```python
# Definimos largos (cte geometrica) y angulos (variables articulares)
l1 = 1
l2 = 1
q1 = pi/6
q2 = -pi/6
```

```python
# Terna inicial
S0 = SE3()
S0
```

```python
# A partir de los datos de la tabla DH, hacemos A01
A01 = SE3.Rz(q1)*SE3.Tx(l1)
A01
```

```python
# Aplicamos A01 a S0, obteniendo S1
S1 = S0 @ A01
S1
```

```python
A12 = SE3.Rz(q2)*SE3.Tx(l2)
A12
```

```python
S2 = S1 @ A12
S2
```

```python
# Ploteamos las tres ternas, en el angulo correspondiente
# El parametro dims es para asegurarnos que los tres ejes tengan la misma escala
# (prueben sacarlo, van a ver que S1 no parece dextrogira)
trplot(S0.A, dims=[-1,5,-1,5,-1,5])
trplot(S1.A)
trplot(S2.A)
```
